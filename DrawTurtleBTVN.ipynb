```python
import turtle
tl = turtle.Turtle()

#Bài 2: Vẽ và tô màu cho hình đa giác đều 10 cạnh
side = 50

tl.pensize(2)
tl.pencolor("pink")
tl.fillcolor("black")
tl.begin_fill()
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.forward(side)
tl.left(36)
tl.end_fill()

#Bài 1: Vẽ hình và tô màu cờ Iceland
height = 180
width = 250

tl.penup()
tl.right(90)
tl.forward(100)
tl.right(90)
tl.forward(30)
tl.left(90)
tl.pendown()
tl.pensize(1)
tl.pencolor("black")
tl.left(90)
tl.forward(width)
tl.right(90)
tl.forward(height)
tl.right(90)
tl.forward(width)
tl.right(90)
tl.forward(height)
tl.right(90)
tl.fillcolor("blue")
tl.begin_fill()
tl.forward(width*0.28)
tl.right(90)
tl.forward(width*0.28)
tl.right(90)
tl.forward(width*0.28)
tl.right(90)
tl.forward(width*0.28)
tl.right(90)
tl.end_fill()
tl.forward(width*0.4)
tl.fillcolor("red")
tl.begin_fill()
tl.right(90)
tl.forward(height*80/180)
tl.left(90)
tl.forward(width*0.6)
tl.right(90)
tl.forward(height/9)
tl.right(90)
tl.forward(width*0.6)
tl.left(90)
tl.forward(height*80/180)
tl.right(90)
tl.forward(width*0.08)
tl.right(90)
tl.forward(height*80/180)
tl.left(90)
tl.forward(width*0.32)
tl.right(90)
tl.forward(height/9)
tl.right(90)
tl.forward(width*0.32)
tl.left(90)
tl.forward(height*80/180)
tl.right(90)
tl.forward(width*0.08)
tl.end_fill()
tl.forward(width*0.04)
tl.fillcolor("blue")
tl.begin_fill()
tl.forward(width*0.56)
tl.right(90)
tl.forward(width*0.28)
tl.right(90)
tl.forward(width*0.56)
tl.right(90)
tl.forward(width*0.28)
tl.end_fill()
tl.right(90)
tl.forward(width*0.56)
tl.right(90)
tl.forward(height*110/180)
tl.fillcolor("blue")
tl.begin_fill()
tl.right(90)
tl.forward(width*0.56)
tl.left(90)
tl.forward(width*0.28)
tl.left(90)
tl.forward(width*0.56)
tl.end_fill()
tl.backward(width)
tl.fillcolor("blue")
tl.begin_fill()
tl.forward(width*0.28)
tl.left(90)
tl.forward(width*0.28)
tl.left(90)
tl.forward(width*0.28)
tl.left(90)
tl.forward(width*0.28)
tl.left(90)
tl.end_fill()

#Bài 2 (cách 2):
tl.penup()
tl.forward(500)
tl.pendown()
tl.circle(100, 360, 10)

turtle.done()
```
