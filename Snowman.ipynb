```python
import turtle
tl = turtle.Turtle()

a = 100
b = 150
#đổ nền đen
turtle.bgcolor("black")
tl.speed(10)
tl.pensize(2)
tl.color("black")
tl.hideturtle()

#vẽ 2 phần thân tròn
tl.fillcolor("white")
tl.begin_fill()
tl.circle(a, 360)
tl.end_fill()
tl.penup()
tl.right(90)
tl.forward(b*2)
tl.left(90)
tl.pendown()
tl.fillcolor("white")
tl.begin_fill()
tl.circle(b, 360)
tl.end_fill()

#vẽ khuy áo
tl.penup()
tl.left(90)
tl.forward(b/2)
tl.dot(20)
tl.forward(b/2)
tl.dot(20)
tl.forward(b/2)
tl.dot(20)

#vẽ mắt
tl.forward((b/2)+(a*2*2/3))
tl.right(90)
tl.forward(30)
tl.dot(20)
tl.backward(60)
tl.dot(20)
tl.forward(30)

#vẽ mũi
tl.penup()
tl.right(90)
tl.forward((a*1/3))
tl.pendown()
tl.fillcolor("orange")
tl.begin_fill()
tl.forward(30)
tl.right(90)
tl.forward(60)
tl.right(150)
tl.forward(((30**2)+(60**2))**0.5)
tl.end_fill()

#vẽ 2 tay
tl.penup()
tl.right(120)
tl.forward(a+b)
tl.right(90)
tl.forward(150)
tl.right(60)
tl.pendown()
tl.pencolor("white")
tl.forward(a)
tl.backward(a/3)
tl.right(30)
tl.forward(a/3)
tl.backward(a/3)
tl.left(120)
tl.forward(a/3)
tl.backward(a/3)
tl.left(90)
tl.forward(a*2/3)
tl.left(60)
tl.penup()
tl.forward(300)
tl.left(60)
tl.pendown()
tl.forward(a)
tl.backward(a/3)
tl.left(30)
tl.forward(a/3)
tl.backward(a/3)
tl.right(120)
tl.forward(a/3)
tl.backward(a/3)
tl.right(90)
tl.forward(a*2/3)
tl.right(60)
tl.penup()
tl.forward(150)
tl.right(90)
tl.forward(350)
tl.left(90)
tl.forward(100)
tl.left(90)
tl.pendown()
tl.fillcolor("blue")
tl.begin_fill()
tl.forward(30)
tl.left(90)
tl.forward(a*2)
tl.left(90)
tl.forward(30)
tl.left(90)
tl.forward(a*2)
tl.end_fill()
tl.backward(a/3)
tl.right(90)
tl.fillcolor("white")
tl.begin_fill()
tl.forward(a/3)
tl.right(90)
tl.forward(a*4/3)
tl.right(90)
tl.forward(a/3)
tl.right(90)
tl.forward(a*4/3)
tl.end_fill()
tl.fillcolor("blue")
tl.begin_fill()
tl.right(90)
tl.forward(a)
tl.right(90)
tl.forward(a*4/3)
tl.right(90)
tl.forward(a*2/3)
tl.right(90)
tl.forward(a*4/3)
tl.end_fill()
tl.penup()
tl.left(90)
tl.forward((a/3)+200)

#vẽ khăn quàng:
tl.pensize(40)
tl.pencolor("red")
tl.left(90)
tl.pendown()
tl.forward(a*4/3)
tl.pensize(30)
tl.right(90)
tl.forward(50)
tl.backward(50)
tl.right(90)
tl.forward(20)
tl.left(90)
tl.forward(30)
tl.penup()

#vẽ miệng cười
tl.right(180)
tl.forward(90)
tl.right(180)
tl.right(90)
tl.forward(80)
tl.left(90)
tl.pendown()
tl.pensize(2)
tl.pencolor("black")
tl.circle(30,180)

turtle.done()
```
